### 3. Univariate Analysis

* Histograms for Numerical Columns

```python
# Set up the plotting style
sns.set(style="whitegrid")

# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot histograms for numerical columns
def plot_histograms(df, cols):
    df[cols].hist(bins=15, figsize=(15, 15), layout=(len(cols)//3 + 1, 3), edgecolor='black')
    plt.tight_layout()
    plt.show()

plot_histograms(df, numerical_cols)
```

* Boxplots to Identify Outliers

```python
# Boxplots for numerical columns
def plot_boxplots(df, cols):
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(cols, 1):
        plt.subplot(len(cols) // 3 + 1, 3, i)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()

plot_boxplots(df, numerical_cols)
```

* Countplots for Categorical Columns

```python
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Plot countplots for categorical columns
def plot_countplots(df, cols):
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(cols, 1):
        plt.subplot(len(cols) // 3 + 1, 3, i)
        sns.countplot(x=df[col], data=df)
        plt.title(f'Countplot of {col}')
        plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

plot_countplots(df, categorical_cols)
```

**Univariate Analysis Observations (Summary):**
1. **Salary:** Right-skewed distribution, with most employees earning between 200,000 and 600,000 thousand. A few outliers represent higher salaries for specialized or senior roles.
   
2. **Experience & Years in Current Role:** Majority have less than 5 years of experience or tenure in their current role, suggesting a younger, early-career workforce with high mobility.
   
3. **Age & Age at Joining:** Most employees are in their mid-20s to early 30s, with many joining their current company in their early 20s. Fewer employees are older or have longer tenures.
   
4. **Performance Rating:** Most employees have high ratings (3-5), indicating a generally strong-performing workforce.
   
5. **Experience (Total):** Majority of employees have between 0-5 years of total experience, indicating a workforce of mostly junior professionals.

6. **MBA Percentage:** Most employees are not MBA graduates, showing that technical roles in this dataset don't heavily favor those with MBAs.
   
7. **Work Experience in the Current Role:** Distribution shows that most employees have only been in their current role for a short period, highlighting quick role transitions.
    
8. **Training Hours:** The distribution is right-skewed, with most employees undergoing fewer training hours, and a few employees receiving significantly higher training time.

2. Relationship Between Gender and Specialization

```python
# Chi-square test for gender and specialization dependence
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square Statistic: {chi2}, p-value: {p}")

if p < 0.05:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```
